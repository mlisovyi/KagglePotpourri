# Composition of Test
The primary goal of this mini-kernel is to probe the leaderboard to assess fractional contribution of different classes to the test sample.  This is important as for a local model evaluation we need to use a split, that will create a validation sample with composition close to our test/submission sample. 

As is demonstrated with a basic EDA in https://www.kaggle.com/mlisovyi/featureengineering-basic-model, `Elevation` is correlated to the target class and at the same time its profile is very different between train and test samples. So we can expect that the class composition differs a lot between the two samples.

The kernel simply creates 7 files each having a fixed label prediction from 1 to 7. One can submit these files and the fraction of corresponding classes will be equal to the returned score.

With this exercise, it was estimated, that the proportion of classes in the test sample is  `0.37053 : 0.49681 : 0.05936 : 0.00103 : 0.01295 : 0.02687 : 0.03242`. So ideally one would create a validation sample that would hold the same mixture.
